```python
# Eric
def univariate_stats(df, freq=12):
    means=df.mean()*freq
    vols=df.std()*(freq**0.5)
    sharpes=means/vols
    return pd.DataFrame({'Mean': means, 'Vol': vols, 'Sharpe': sharpes})
```

```python
# Eric
import statsmodels.api as sm
def simple_ols(y, X):
    model=sm.OLS(y, sm.add_constant(X)).fit()
    alpha=model.params[0]
    beta=model.params[1]
    info=model.params[0]/model.resid.std()
    treynor=y.mean()/beta
    rsquared=model.rsquared
    return pd.DataFrame([alpha, beta, info, treynor, rsquared], index=['Alpha', 'Beta', 'Info Ratio', 'Treynor Ratio', 'R-Squared'])
```

```python
# Eric
def multi_ols(y,X):
    model=sm.OLS(y, sm.add_constant(X)).fit()
    alpha=model.params[0]
    info=model.params[0]/model.resid.std()
    rsquared=model.rsquared
    return {'Alpha': alpha, 'Info Ratio': info, 'R-Squared': rsquared, 'Coeffs': model.params[1:]}
```

```python
# Eric
mu=factors.mean()
sigma=factors.cov()
w=np.linalg.inv(sigma)@mu
w_tan=w/w.sum()
pd.DataFrame(w_tan, index=mu.index, columns=['Weight'])
```

Most important factors are CMA and RMW, and least important is HML. 

The factors with low mean returns are SMB, HML, and CMA, and CMA has a large weight in the tangency portfolio, so it's pretty useful.

```python
# Eric
mu=factors[['MKT', 'SMB', 'HML', 'UMD']].mean()
sigma=factors[['MKT', 'SMB', 'HML', 'UMD']].cov()
w=np.linalg.inv(sigma)@mu
w_tan=w/w.sum()
pd.DataFrame(w_tan, index=mu.index, columns=['Weight'])
```

This time MKT and HML have high weights, and SMB has low weight. 

Overall, HML and CMA have a bit of redundancy with each other. SMB does not seem to be an important factor according to the tangency weights, and it also has a low mean return. The other factors are all pretty important.
